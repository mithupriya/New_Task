```python
from bs4 import BeautifulSoup
import lxml
import requests
import pandas as pd
import numpy as np
```

```python
url="https://www.pathologyoutlines.com/"
```

```python
r=requests.get(url)
htmlContent=r.content
print(htmlContent)
```

```python
soup=BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify)
```

```python
title=soup.title
print(type(soup))
print(type(title))
print(type(title.string))
```
